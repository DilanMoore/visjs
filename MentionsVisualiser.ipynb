```python
import json
import pandas as pd
import numpy as np

import holoviews as hv
from holoviews import opts, dim
import holoviews.plotting.bokeh
hv.extension('bokeh')
hv.output(size=200)
```

```python
with open('nonVar_session_restructured.js', 'r') as f:
    sessionData = json.load(f)
```

```python
playerList = []
for message in sessionData[0]['messages']:
    player = message['player']
    if player not in playerList:
        playerList.append(player)
```

```python
mentionDict = dict.fromkeys(playerList)
for player in playerList:
    mentionDict[player] = []
```

```python
for message in sessionData[0]['messages']:
    if len(message['mentions']) != 0:
        newMentions = message['mentions']
        player = message['player']
        
        if mentionDict[player] == None:
            mentionDict[player] = newMentions
        else:
            existingMentions = mentionDict[player]
            mentionDict[player] = existingMentions + newMentions
```

```python
sourceList = []
targetList = []
valueList = []
for key in mentionDict.keys():
    for item in mentionDict[key]:
        sourceList.append(key)
        targetList.append(item)
        valueList.append(1)
    
mentionsDF = pd.DataFrame({"source" : sourceList, "target" : targetList, "value" : valueList, "name" : sourceList})
```

```python
mentionsDF
```

```python
chord = hv.Chord(mentionsDF)

chord.opts(
    opts.Chord(cmap='Category20', edge_cmap='Category20', edge_color=dim('source').str(), 
               labels='name', node_color=dim('index').str()))

hv.save(chord, "chordPlot.html")
```

```python
# make plots interactive

%matplotlib nbagg


import seaborn as sns
import seaborn.objects as so
import matplotlib.pyplot as plt
import pandas as pd
import plotly
```

```python
messageData = pd.DataFrame(sessionData[0]['messages'])
```

```python
messageData["messageCount"] = 1
messageData["charCount"] = messageData["message"].str.len()

messageData["rawCharCount"] = messageData["message"].str.len()
```

```python
messageData["totalActivity"] = messageData.rolling(10).mean()["charCount"]

messageData["messageCount"] = messageData.groupby(by = "player").cumsum()["messageCount"]

messageData["charCount"] = messageData.groupby(by = "player").cumsum()["charCount"]
```

```python
p = so.Plot(messageData, "datetime", "charCount", color = "player")#.facet("player")
p = p.add(so.Area())
p.save("snsPlot.png")
```
